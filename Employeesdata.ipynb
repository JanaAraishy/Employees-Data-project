# Clean Data

```python
df=df.drop_duplicates()
df
```

```python
df=df.dropna(subset=["Full Name","Department","Job Title"])
```

```python
cat_cols = [
    "Gender",
    "Ethnicity", "Country", "City"
]

for col in cat_cols :
  df[col]=df[col].fillna(df[col].mode()[0])
```

```python
num_cols=["Age","Annual Salary","Bonus %"]

for col in num_cols:
  df[col]=df[col].fillna(df[col].median())
```
